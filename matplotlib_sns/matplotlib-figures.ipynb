```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
a = np.linspace(0,10,11)
```

```python
b = a ** 4
```

```python
x = np.arange(0,10)
```

```python
y = 2 * x
```

```python
fig,axes = plt.subplots(nrows=2,ncols=2)

axes[0][1].plot(x,y)
axes[1][0].plot(a,b)

fig.suptitle('elly pro')
fig.tight_layout()
```

```python
axes
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

# Customizing lines
ax.plot(x,y,label='y=2*x',c='magenta',lw=5,ls='dashed')

# Customizing markers
ax.plot(x,x+1,label='y=x+1',c='red',
       lw=2,marker='o',ms=10,
       mfc='orange',mew=2,mec='blue'
       )
# Legend shows
ax.legend(loc='best')

# Grid
plt.grid()

plt.show()
```
