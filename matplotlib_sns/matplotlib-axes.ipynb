```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
a = np.linspace(0,10,11)
```

```python
a
```

```python
b = a**4
```

```python
b
```

```python
x = np.arange(0,10)
```

```python
x
```

```python
y = 2*x
```

```python
y
```

```python
fig = plt.figure()

# Large axes
axes1 = fig.add_axes([0,0,1,1])
axes1.set_title('a=b**4')
axes1.set_xlim(0,8)
axes1.set_ylim(0,8000)
axes1.set_xlabel('a')
axes1.set_ylabel('b')
axes1.plot(a,b)

# Small axes
axes2 = fig.add_axes([0.2,0.2,0.5,0.5])
axes2.plot(x,y)

plt.show()
```

```python
fig = plt.figure(figsize=(10,5),dpi=100)

axes = fig.add_axes([0,0,1,1])

axes.plot(a,b)

fig.savefig('power4.png',bbox_inches='tight')
```
